In class assignment

Find most recent & the earliest travelled date and mean distance traveled for each start city

```python
df["START_DATE*"] = pd.to_datetime(df["START_DATE*"])
df.dtypes
```

```python
 df.groupby("START*")["START_DATE*","MILES*"].agg([("min","max"),"mean"]).head(20)
```

```python
res =  df.groupby("START*").agg({"MILES*": ["mean"], "START_DATE*": ["min","max"]}).head(20)
res.reset_index(inplace = True)
res
```

```python
res.columns = ["START_CITY", "AVG_DIST", "EARLIEST_TRIP","RECENT_TRIP"]
res
res.to_csv("new_uber.csv")
```
